```python
class Human:
    
    population = 0
    is_seq = 101
    database = []
    
    
    def __init__(self, name, age, is_alive = True):
        print('Creating new object')
        
        self.name = name
        self.age = age
        self.is_alive = is_alive
        
        self.id = Human.is_seq
        Human.population+=1
        Human.is_seq+=1
        
        Human.database.append(self)
        
        
    def introduce(self):
        print('Hi my name is {} and i am {} year old'.format(self.name,self.age))
        
    def die(self):
        if (self.is_alive):
            print(self.name, 'is dying...')
            self.is_alive = False
            Human.population-=1
        else:
            print(self.name, 'is already dead..')
            
```

```python
def Lcm(a,b):
    """return Lcm of two numbers"""
    
    max_num = (a if a>b else b)
    
    while True:
        if (max_num%a==0) and (max_num%b==0):
            return max_num
        else:
            max_num+=1
```

```python
df = pd.read_csv('movie_meta.csv')
```

```python
def preprocess(ele):
    return len(ele.strip())
```

```python
df['movie_title_len'] = df['movie_title'].apply(preprocess)
```

```python
df.columns
```

```python
plt.hist(df['movie_title_len'], bins = 50)
plt.xticks(range(0,100,5))
plt.xlabel('movie_title_length')
plt.ylabel('frequency')
plt.show()
```

```python
freq_dic = {}
for i in df['movie_title_len']:
    if freq_dic.get(i) is None:
        freq_dic[i] = 1
    else:
        freq_dic[i] += 1
    
```

```python
x = freq_dic.keys()
```

```python
y = freq_dic.values()
```

```python
plt.scatter(x,y)
plt.xlabel('movie_title_length')
plt.xticks(range(0,60,5))
plt.ylabel('frequency')
plt.show()
plt.show()
```
