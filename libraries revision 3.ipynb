```python
df = pd.read_csv('https://raw.githubusercontent.com/coding-minutes/python-data-science-mastercourse/main/06%20Matplotlib/movie_meta.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df['movie_title_length'] = df['movie_title'].str.strip().apply(len)
```

```python
df.columns
```

```python
plt.hist(df['movie_title_length'], bins=50)
plt.xticks(range(0,100,5))
plt.xlabel('movie_title_length')
plt.ylabel('frequency')
plt.show()
```

```python
freq_title = {}

for i in df['movie_title_length']:
    if i not in freq_title:
        freq_title[i] = 1
    else:
        freq_title[i] += 1
```

```python
freq_title
```

```python
x = freq_title.keys()
```

```python
y = freq_title.values()
```

```python
plt.scatter(x,y)
plt.xticks(range(0,60,5))
plt.xlabel('movie_title_len')
plt.ylabel('frequency')
plt.show()
```
