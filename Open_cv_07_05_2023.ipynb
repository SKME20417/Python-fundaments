```python
rows, cols = img_rgb.shape[:-1]
```

```python
m = cv2.getRotationMatrix2D((rows/2, cols/2), 360, 0.90)
```

```python
i = cv2.warpAffine(img_rgb,m,(cols,rows))
```

```python
plt.imshow(i)
```

```python
M = np.float32([[1,0,50],[0,1,20]])
```

```python
i = cv2.warpAffine(img_rgb, M, (cols, rows))
```

```python
plt.imshow(i)
```

```python
M = np.float32([[1,0,-50],[0,1,-20]])
```

```python
i = cv2.warpAffine(img_rgb, M, (cols, rows))
```

```python
plt.imshow(i)
```
