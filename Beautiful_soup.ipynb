```python
import requests
import bs4
import pandas as pd
import re
```

```python
res = requests.get("https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_and_their_capitals_in_native_languages")
```

```python
res.status_code
```

```python
text = res.content
```

```python
text
```

```python
soup = bs4.BeautifulSoup(text)
```

```python
type(soup)
```

```python
soup.title
```

```python
soup.h1.text
```

```python
soup.find('h1')
```

```python
soup.h1
```

```python
all_tables = soup.find_all('table', attrs={'class':'wikitable'})
```

```python
len(all_tables)
```

```python
B_table = all_tables[1].find_all('tr')[1:]
```

```python
len(B_table)
```

```python
all_tds = B_table[0].find_all('td')
```

```python
len(all_tds)
```

```python
all_tds[0].img.attrs['src']
```

```python
country = all_tds[0].text
```

```python
capital = all_tds[1].text
```

```python
language = all_tds[-1].text
```

```python
print(country, capital, language)
```

```python
df = pd.DataFrame({}, columns=['Country', 'Capital', 'Language'])
```

```python
df
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python

for one_table in all_tables:
    all_rows= one_table.find_all('tr')[1:]
    for tr in all_rows:
        all_tds = tr.find_all('td')

        country = all_tds[0].text
        capital = all_tds[1].text
        language = all_tds[-1].text

        ele = {'Country': country, 'Capital' : capital, 'Language' : language}
        
        df = df.append(ele, ignore_index=True)
        
        print(country,"==",capital,"==",language)
        print()
    
```

```python
df.shape
```

```python
df.head()
```

```python
df.to_csv('countries.csv', index=False)
```

```python
df['Country'] = df.Country.apply(lambda x:re.sub('\[.*\]', "", x.strip()))
```

```python
res = requests.get('https://www.passiton.com/inspirational-quotes?page=2')
```

```python
data = res.content
```

```python
soup = bs4.BeautifulSoup(data)
```

```python
box = soup.find("div", attrs={'id':"all_quotes"})
```

```python
imgs = box.find_all("img")
```

```python
for idx,i in enumerate(imgs, start=34):
    img_url = i.attrs['src']
    res = requests.get(img_url)
    bin_data = res.content
    
    with open(f"quotes\quotes{idx}.jpg", 'wb') as f:
        f.write(bin_data)
    
```
